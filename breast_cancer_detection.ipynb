```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,f1_score
```

```python
data_set=pd.read_csv(r"C:\Users\gangu\Downloads\breast cancer.csv")
data_set.head()
```

```python
data_set.shape
```

```python
data_set.info()
```

```python
data_set.replace({'diagnosis':{'M':0,'B':1}},inplace=True)
```

```python
data_set.head()
```

```python
data_set['diagnosis'].value_counts()
```

```python
x=data_set.drop(columns='diagnosis')
x.head()
```

```python
y=data_set['diagnosis']
y
```

```python
scaler=StandardScaler()
scaler
```

```python
x=scaler.fit_transform(x)
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
model=KNeighborsClassifier()
```

```python
model
```

```python
model.fit(x_train,y_train)
```

```python
train_data_pre=model.predict(x_train)
acc_score=accuracy_score(y_train,train_data_pre)
print('accuracy score of training data is:',acc_score)
cm_train_data=confusion_matrix(y_train,train_data_pre)
print(cm_train_data)
```

```python
test_data_pre=model.predict(x_test)
acc_score_test=accuracy_score(y_test,test_data_pre)
print('accuracy score of testing data is:',acc_score_test)
cm=confusion_matrix(y_test,test_data_pre)
print(cm)
```
